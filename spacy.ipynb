# Part of speech (POS)

```python
!pip install beautifultable
```

```python
from beautifultable import BeautifulTable

table = BeautifulTable()
```

```python
text = "Apple is looking at buying U.K. startup for $1 billion"
doc = nlp(text)

table.columns.header = ["TEXT", "POS", "TAG", "DEP", "LEMMA", "SHAPE", "ALPHA", "STOP"]

for token in doc:
  table.rows.append([token.text, token.pos_, token.tag_, token.dep_, token.lemma_, token.shape_, token.is_alpha, token.is_stop])

print(table)
```

```python
from spacy import displacy
```

```python
options = {"compact":True, "distance": 150, "bg":"#0095b6", "color":"#DDDDDD", "font":"Source Sans Pro"}
displacy.render(doc, style="dep", jupyter=True, options=options)
```

# Sentence Boundary Detection

```python
para = "If we are unable to complete our initial business combination within 33 months from the closing of this offering, unless we extend such period via sponsor commitment, we will redeem 100% of the public shares at a per-share price equal to the aggregate amount on deposit in the trust account, including interest earned thereon (less taxes payable and up to $65,000 of interest to pay dissolution expenses), divided by the number of then outstanding public shares. Shareholders could face liability if distributions are deemed unlawful."
```

```python
doc = nlp(para)
```

```python
list(doc)
```

```python
sents = list(doc.sents)

len(sents)
```

```python
for sent in sents:
  print(sent)
```

# **Stop Words**

```python
stopwords = spacy.lang.en.stop_words.STOP_WORDS
```

```python
stopwords
```

```python
for sent in sents:
  print(sent.text)
  for token in sent:
    if token.text not in stopwords:
      print(token)
  break
```

# Word Frequency

```python
text = "If we anticipate that we may not be able to consummate our initial business combination within 12 months from the effective date of this registration statement, we may, but are not obligated to, extend the period of time to consummate a business combination two times by an additional three months each time (for a total of up to 18 months to complete a business combination), provided that our sponsor or designee must deposit into the trust account for each three months extension, $600,000, or $690,000 if the underwriters’ over-allotment option is exercised in full ($0.10 per share in either case), up to an aggregate of $1,200,000 or $1,380,000 if the underwriters’ over-allotment option is exercised in full, on or prior to the date of the applicable deadline."
```

```python
from collections import Counter
```

```python
word_freq = Counter(text.split())
word_freq
```

```python
word_freq.most_common(5)
```

```python
doc = nlp(text)
```

```python
words = [token.text for token in doc if not token.is_stop and not token.is_punct]
```

```python
word_freq = Counter(words)
word_freq
```
